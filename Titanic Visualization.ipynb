```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train=pd.read_csv('Datasets/Titanic/train.csv')
```

```python
test=pd.read_csv('Datasets/Titanic/test.csv')
```

# Exploratory Data Analysis

## Missing Data

```python
#We can use seaborn to create a simple heatmap to see where we are missing data!

sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
sns.heatmap(test.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

Roughly 20 percent of the Age data is missing. We'll fill this data with other features.

```python
sns.countplot(x='Survived',data=train,palette='RdBu_r')
```

Our Dataset isn't balanced as there are way more data points with the people perishing as compared to surving

```python
sns.countplot(x='Survived',hue='Sex',data=train,palette='RdBu_r')
```

way more women survived than men

```python
sns.countplot(x='Survived',hue='Pclass',data=train,palette='rainbow')
```

Most people in 3rd class didn't survive, more than half the people in 1st class did which is more than the ratio of people who survived in the entire data set. In 2nd class, it's a equal ratio.

```python
sns.displot(train['Age'].dropna(),kde=False,color='darkred',bins=30)
```

```python
sns.displot(test['Age'].dropna(),kde=False,color='darkred',bins=30)
```

Majority of people on board were in their 20s

```python
sns.countplot(x='SibSp',data=train)
```

```python
sns.countplot(x='SibSp',data=test)
```

Majority of the people on board the titanic, didn't have siblings or spouses on board

```python
train['Fare'].hist(color='green',bins=40,figsize=(8,4))
```

```python
test['Fare'].hist(color='green',bins=40,figsize=(8,4))
```

Majority of the people paid between 0-12 dollars for their tickets, which suggests most people were in the 3rd class

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=train,palette='winter')
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=test,palette='winter')
```
